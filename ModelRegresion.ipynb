```python
df= pd.read_csv("data4.csv")
```

```python
columnas_a_eliminar = ['Unnamed: 0.1', 'Unnamed: 0','Unnamed: 35','Unnamed: 36']
```

```python
df = df.drop(columnas_a_eliminar, axis=1)
```

```python
df.head(5)
```

```python
Vars_Indep= df[['monto_financiado']]
Var_Dep= df['costo_total']
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

```python
model.fit(X=Vars_Indep, y=Var_Dep)
```

```python
model.__dict__
```

```python
y_pred= model.predict(X=df[['monto_financiado']])
y_pred
```

```python
df.insert(0, 'Predicciones', y_pred)
df
```

```python
coef_Deter=model.score(X=Vars_Indep, y=Var_Dep)
coef_Deter
```

```python
coef_Correl=np.sqrt(coef_Deter)
coef_Correl
```

```python
sns.scatterplot(x='monto_financiado', y='costo_total', color="blue", data=df)
sns.scatterplot(x='monto_financiado', y='Predicciones', color="red", data=df)
```
